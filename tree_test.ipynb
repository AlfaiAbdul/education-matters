```python

import warnings
warnings.simplefilter('ignore')

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the csv file into a pandas DataFrame

edu_df = pd.read_csv('consol_data.csv')
edu_df.head()
```

```python
# drop PrivacySuppressed rows
edu_df = edu_df.replace('PrivacySuppressed', np.nan).dropna()
edu_df.head()
```

```python
# List columns of the dataframe
edu_df.columns
```

```python
# get data types of columns
#edu_df.dtypes
```

```python
# convert the columns into numeric
edu_df['MN_EARN_WNE_P10'] = pd.to_numeric(edu_df['MN_EARN_WNE_P10'], errors='coerce')
edu_df['FEMALE'] = pd.to_numeric(edu_df['FEMALE'], errors='coerce')
```

```python
# get data types of columns
#edu_df.dtypes
```

```python
# convert the numerical earnings  to bins to give it as an output to the decision tree


### #bins = [0, 24999, 49999, 74999, 99999, 124999, 149999]
#group_names = ['0-25','25-50','50-75','75-100','100-125', '125-150']

#bins = [0, 42999, 50999, 60999, 74999, 99999, 149999]
#group_names = ['0-43000','43000-51000','51000-61000','61000-75000','75000-100000', '100000-150000']

# bins = [0, 29999, 34999, 39999, 44999, 49999, 54999, 59999, 64999, 69999, 74999, 79999, 84999, 89999, 94999, 99999, \
#         104999, 109999, 149999]
# group_names = ['Below 30000','30-35', '35-40', '40-45', '45-50', '50-55', '55-60', '60-65', '65-70',\
# '70-75', '75-80', '80-85', '85-90', '90-95', '95-100', '100-105', '105-110','110 and Above']

#Bins based on k-means clustering
# bins = [0, 42000, 50000, 60000, 75000, 150000]
# group_names = ['Below 42000', '42-50', '50-60',  '60-75', '75 and Above']

bins = [0, 29999, 49999, 69999, 89999, 109999, 129999, 149999]
group_names = ['Below 30K', '30K-50K', '50K-70K',  '70K-90K', '90K-110K',  '110K-130K', '130K-150K']


# Categorize the existing players using the age bins
edu_df["Sal Ranges"] = pd.cut(edu_df["MN_EARN_WNE_P10"], bins, labels=group_names)
#edu_df["Sal Ranges"]
```

```python
target = edu_df["Sal Ranges"]
```

```python
data = edu_df[["PCT_WHITE", "PCT_BLACK", "PCT_ASIAN", "PCT_HISPANIC", 
                "SAT_AVG", "TUITIONFEE_IN", 'AGE_ENTRY', 'MEDIAN_HH_INC',
               'POVERTY_RATE', 'FEMALE'
               #"POVERTY_RATE","UNEMP_RATE",'FIRST_GEN', 'FAMINC', 
               #'C150_4_WHITE','C150_4_BLACK', 'C150_4_HISP', 'C150_4_ASIAN', 'PCIP01', 'PCIP03',
               #'PCIP04', 'PCIP05', 'PCIP09', 'PCIP10', 'PCIP11', 'PCIP12', 'PCIP13',
               #'PCIP14', 'PCIP15', 'PCIP16', 'PCIP19', 'PCIP22', 'PCIP23', 'PCIP24',
               #'PCIP25', 'PCIP26', 'PCIP27', 'PCIP29', 'PCIP30', 'PCIP31', 'PCIP38',
               #'PCIP39', 'PCIP40', 'PCIP41', 'PCIP42', 'PCIP43', 'PCIP44', 'PCIP45',
               #'PCIP46', 'PCIP47', 'PCIP48', 'PCIP49', 'PCIP50', 'PCIP51', 'PCIP52',
               #'PCIP54'            
              ]]
feature_names = data.columns
data.head()
```

# Using RandomForestClassifier to get the list of important features

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
X_train, X_test, y_train, y_test = train_test_split(data, target, stratify=target, 
                                                    test_size=0.25, random_state=42)
```

# Using DecisionTreeClassifier with 4 most important features

```python
from sklearn.model_selection import train_test_split
from sklearn import tree

# define our X and y
data = edu_df[["SAT_AVG", "TUITIONFEE_IN", 'AGE_ENTRY', 'MEDIAN_HH_INC']]
target = edu_df["Sal Ranges"] 

# use stratify to sample the data evenly
X_train, X_test, y_train, y_test = train_test_split(data, target, stratify=target, 
                                                    test_size=0.25, random_state=42)


clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# pass a random value to see the model's prediction
clf.predict(np.reshape([ 850, 93666, 20.27, 49720], (1,4)))
```

# Visualize the Decision tree

```python
import graphviz
import pydotplus

dot_data =tree.export_graphviz(clf,out_file=None,feature_names=["SAT", "TUITIONFEE", 'AGE', 'HH_INC'],  
                         class_names=['Below 30000', '30-50', '50-70',  '70-90', '90-110',  '110-130', '130-150'],  
                         filled=True, rounded=True,  
                         special_characters=True) 

# save the decision tree as an image file
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_png('earnings_tree.png')
graph = graphviz.Source(dot_data)  
graph
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
